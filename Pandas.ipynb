<h1>isin()</h1>

```python
df2['batsman'].value_counts()
```

```python
#most destractive death over(16-20) batsman in ipl hitory
#who played minmum 200 ball
#basies of most strike rate=(no. of runs/no. of balls)/100
```

```python
#first find players who played 200 ball among 16-20 overs
deadth_over_batsman=df2[df2['over']>15].groupby('batsman')['batsman_runs'].count()
deadth_over_batsman
```

```python
batsman_200=deadth_over_batsman[deadth_over_batsman>200].index.tolist()
batsman_200
```

```python
#now we need these 200 player strike rate  and rank
new_df2=df2[df2['batsman'].isin(batsman_200)]
new_df2
```

```python
#calculate strike rate=(no. of runs/no. of balls)*100
runs=new_df2.groupby('batsman')['batsman_runs'].sum()
runs
```

```python
balls=new_df2.groupby('batsman')['batsman_runs'].count()
balls
```

```python
strike_rate=(runs/balls)*100
strike_rate.sort_values(ascending=False).head(5)
```
