#Heatmap(Cat-Cat)
## to show relation between two Number Catagorcal data

```python
#prottok Pclass a koto golo lok jibito ase ar koto golo lok maragase
# sns.barplot(data=ti,x='Pclass',y='Survived')
pd.crosstab(ti['Pclass'],ti['Survived'])
```

```python
ti.groupby('Pclass')['Survived'].mean()#catagorical-catagorical
```
