

## üìå **Principal Component Analysis (PCA) ‚Äì ‡¶™‡ßÇ‡¶∞‡ßç‡¶£ ‡¶ó‡¶æ‡¶£‡¶ø‡¶§‡¶ø‡¶ï ‡¶ì ‡¶™‡ßç‡¶∞‡¶æ‡¶ï‡¶ü‡¶ø‡¶ï‡ßç‡¶Ø‡¶æ‡¶≤ ‡¶¨‡ßç‡¶Ø‡¶æ‡¶ñ‡ßç‡¶Ø‡¶æ**

### Read Curse of Dimensionality

### üéØ **PCA ‡¶ï‡ßÄ?**

PCA (Principal Component Analysis) ‡¶è‡¶ï‡¶ü‡¶ø ‡¶°‡¶æ‡¶á‡¶Æ‡ßá‡¶®‡¶∂‡¶® ‡¶∞‡¶ø‡¶°‡¶æ‡¶ï‡¶∂‡¶® ‡¶ü‡ßá‡¶ï‡¶®‡¶ø‡¶ï, ‡¶Ø‡¶æ‡¶∞ ‡¶Æ‡¶æ‡¶ß‡ßç‡¶Ø‡¶Æ‡ßá ‡¶Ü‡¶Æ‡¶∞‡¶æ ‡¶â‡¶ö‡ßç‡¶ö-‡¶°‡¶æ‡¶á‡¶Æ‡ßá‡¶®‡¶∂‡¶®‡ßá‡¶∞ ‡¶°‡ßá‡¶ü‡¶æ‡¶ï‡ßá ‡¶è‡¶Æ‡¶® ‡¶è‡¶ï‡¶ü‡¶ø ‡¶®‡¶ø‡¶Æ‡ßç‡¶®-‡¶°‡¶æ‡¶á‡¶Æ‡ßá‡¶®‡¶∂‡¶®‡¶æ‡¶≤ ‡¶∏‡ßç‡¶™‡ßá‡¶∏-‡¶è ‡¶∞‡ßÇ‡¶™‡¶æ‡¶®‡ßç‡¶§‡¶∞ ‡¶ï‡¶∞‡¶ø, ‡¶Ø‡ßá‡¶ñ‡¶æ‡¶®‡ßá **‡¶∏‡¶∞‡ßç‡¶¨‡¶æ‡¶ß‡¶ø‡¶ï ‡¶≠‡ßç‡¶Ø‡¶æ‡¶∞‡¶ø‡¶Ø‡¶º‡ßá‡¶®‡ßç‡¶∏** ‡¶ß‡¶∞‡ßá ‡¶∞‡¶æ‡¶ñ‡¶æ ‡¶Ø‡¶æ‡¶Ø‡¶º‡•§

---

### üßÆ **Step 1: ‡¶°‡ßá‡¶ü‡¶æ Standardization**

#### üî¢ **‡¶ó‡¶£‡¶ø‡¶§:**

$$
Z = \frac{X - \mu}{\sigma}
$$

* $X$: ‡¶Æ‡ßÇ‡¶≤ ‡¶°‡ßá‡¶ü‡¶æ ‡¶Æ‡ßç‡¶Ø‡¶æ‡¶ü‡ßç‡¶∞‡¶ø‡¶ï‡ßç‡¶∏ (‡¶Ü‡¶ï‡¶æ‡¶∞: $n \times p$)
* $\mu$: ‡¶™‡ßç‡¶∞‡¶§‡¶ø‡¶ü‡¶ø ‡¶´‡¶ø‡¶ö‡¶æ‡¶∞‡ßá‡¶∞ ‡¶ó‡¶°‡¶º
* $\sigma$: ‡¶™‡ßç‡¶∞‡¶§‡¶ø‡¶ü‡¶ø ‡¶´‡¶ø‡¶ö‡¶æ‡¶∞‡ßá‡¶∞ ‡¶∏‡ßç‡¶ü‡ßç‡¶Ø‡¶æ‡¶®‡ßç‡¶°‡¶æ‡¶∞‡ßç‡¶° ‡¶°‡ßá‡¶≠‡¶ø‡¶Ø‡¶º‡ßá‡¶∂‡¶®

#### üé≤ **‡¶â‡¶¶‡¶æ‡¶π‡¶∞‡¶£:**

| Sample | X‚ÇÅ | X‚ÇÇ |
| ------ | -- | -- |
| 1      | 2  | 3  |
| 2      | 4  | 1  |
| 3      | 6  | 5  |

‡¶ó‡¶°‡¶º:

$$
\mu_{X‚ÇÅ} = 4, \quad \mu_{X‚ÇÇ} = 3
$$

‡¶∏‡ßç‡¶ü‡ßç‡¶Ø‡¶æ‡¶®‡ßç‡¶°‡¶æ‡¶∞‡ßç‡¶°‡¶æ‡¶á‡¶ú‡¶° ‡¶°‡ßá‡¶ü‡¶æ:

$$
Z = \begin{bmatrix}
\frac{2-4}{\sigma_{X‚ÇÅ}} & \frac{3-3}{\sigma_{X‚ÇÇ}} \\
\frac{4-4}{\sigma_{X‚ÇÅ}} & \frac{1-3}{\sigma_{X‚ÇÇ}} \\
\frac{6-4}{\sigma_{X‚ÇÅ}} & \frac{5-3}{\sigma_{X‚ÇÇ}}
\end{bmatrix}
$$

---

### üßæ **Step 2: ‡¶ï‡ßã‡¶≠‡ßç‡¶Ø‡¶æ‡¶∞‡¶ø‡¶Ø‡¶º‡ßá‡¶®‡ßç‡¶∏ ‡¶Æ‡ßç‡¶Ø‡¶æ‡¶ü‡ßç‡¶∞‡¶ø‡¶ï‡ßç‡¶∏ ‡¶®‡¶ø‡¶∞‡ßç‡¶£‡¶Ø‡¶º**

#### üî¢ **‡¶ó‡¶£‡¶ø‡¶§:**

$$
\Sigma = \frac{1}{n} Z^T Z
$$

* $\Sigma$: ‡¶ï‡ßã‡¶≠‡ßç‡¶Ø‡¶æ‡¶∞‡¶ø‡¶Ø‡¶º‡ßá‡¶®‡ßç‡¶∏ ‡¶Æ‡ßç‡¶Ø‡¶æ‡¶ü‡ßç‡¶∞‡¶ø‡¶ï‡ßç‡¶∏ (‡¶Ü‡¶ï‡¶æ‡¶∞: $p \times p$)

#### üéØ **‡¶â‡¶¶‡¶æ‡¶π‡¶∞‡¶£:**

$$
\Sigma = \begin{bmatrix}
\text{Var}(Z‚ÇÅ) & \text{Cov}(Z‚ÇÅ,Z‚ÇÇ) \\
\text{Cov}(Z‚ÇÇ,Z‚ÇÅ) & \text{Var}(Z‚ÇÇ)
\end{bmatrix}
$$

---

### üîç **Step 3: ‡¶Ü‡¶á‡¶ó‡ßá‡¶®‡¶≠‡ßç‡¶Ø‡¶æ‡¶≤‡ßÅ ‡¶ì ‡¶Ü‡¶á‡¶ó‡ßá‡¶®‡¶≠‡ßá‡¶ï‡ßç‡¶ü‡¶∞ ‡¶®‡¶ø‡¶∞‡ßç‡¶£‡¶Ø‡¶º**

#### üî¢ **‡¶ó‡¶£‡¶ø‡¶§:**

$$
\Sigma \mathbf{v} = \lambda \mathbf{v}
$$

* $\lambda$: Eigenvalue
* $\mathbf{v}$: Eigenvector

#### üßÆ **‡¶ß‡¶æ‡¶™‡¶∏‡¶Æ‡ßÇ‡¶π:**

1. Characteristic Equation:

   $$
   \det(\Sigma - \lambda I) = 0
   $$
2. $(\Sigma - \lambda I) \mathbf{v} = 0$ ‡¶∏‡¶Æ‡¶æ‡¶ß‡¶æ‡¶® ‡¶ï‡¶∞‡ßá $\mathbf{v}$ ‡¶®‡¶ø‡¶∞‡ßç‡¶£‡¶Ø‡¶º

#### üß™ **‡¶â‡¶¶‡¶æ‡¶π‡¶∞‡¶£:**

$$
\Sigma = \begin{bmatrix} 2 & 1 \\ 1 & 2 \end{bmatrix}
\Rightarrow \lambda = 1, 3
$$

Eigenvectors:

* $\lambda = 3$: $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$
* $\lambda = 1$: $\begin{bmatrix} -1 \\ 1 \end{bmatrix}$

---

### üìä **Step 4: Principal Components ‡¶®‡¶ø‡¶∞‡ßç‡¶¨‡¶æ‡¶ö‡¶®**

#### üî¢ **‡¶ó‡¶£‡¶ø‡¶§:**

$$
\text{Cumulative Variance} = \frac{\sum_{i=1}^k \lambda_i}{\sum_{i=1}^p \lambda_i}
$$

#### üìå **‡¶â‡¶¶‡¶æ‡¶π‡¶∞‡¶£:**

$$
\lambda_1 = 3, \lambda_2 = 1 \Rightarrow \text{Total Variance} = 4
\Rightarrow PC1 = 75\%, \quad PC1 + PC2 = 100\%
$$

---

### üîÑ **Step 5: ‡¶°‡ßá‡¶ü‡¶æ ‡¶ü‡ßç‡¶∞‡¶æ‡¶®‡ßç‡¶∏‡¶´‡¶∞‡ßç‡¶Æ‡ßá‡¶∂‡¶®**

#### üî¢ **‡¶ó‡¶£‡¶ø‡¶§:**

$$
Y = Z W_k
$$

* $W_k$: ‡¶ü‡¶™ $k$ ‡¶Ü‡¶á‡¶ó‡ßá‡¶®‡¶≠‡ßá‡¶ï‡ßç‡¶ü‡¶∞‡ßá‡¶∞ ‡¶Æ‡ßç‡¶Ø‡¶æ‡¶ü‡ßç‡¶∞‡¶ø‡¶ï‡ßç‡¶∏
* $Y$: ‡¶ü‡ßç‡¶∞‡¶æ‡¶®‡ßç‡¶∏‡¶´‡¶∞‡ßç‡¶Æ‡¶° ‡¶°‡ßá‡¶ü‡¶æ

#### üß™ **‡¶â‡¶¶‡¶æ‡¶π‡¶∞‡¶£:**

$$
Y = Z \begin{bmatrix} 1 \\ 1 \end{bmatrix}
$$

## üíª **Python Code: PCA ‡¶¨‡¶æ‡¶∏‡ßç‡¶§‡¶¨‡¶æ‡¶Ø‡¶º‡¶®**

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Step 1: Raw Data
X = np.array([[2, 3], [4, 1], [6, 5]])

# Step 2: Standardization
scaler = StandardScaler()
Z = scaler.fit_transform(X)

# Step 3: PCA
pca = PCA(n_components=1)
X_pca = pca.fit_transform(Z)

# Output
print("üîÑ ‡¶ü‡ßç‡¶∞‡¶æ‡¶®‡ßç‡¶∏‡¶´‡¶∞‡ßç‡¶Æ‡¶° ‡¶°‡ßá‡¶ü‡¶æ:\n", X_pca)
print("üìà ‡¶è‡¶ï‡ßç‡¶∏‡¶™‡ßç‡¶≤‡ßá‡¶á‡¶®‡ßç‡¶° ‡¶≠‡ßç‡¶Ø‡¶æ‡¶∞‡¶ø‡¶Ø‡¶º‡ßá‡¶®‡ßç‡¶∏ ‡¶∞‡ßá‡¶∂‡¶ø‡¶ì:", pca.explained_variance_ratio_)
```

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as ex
````

````python
df=pd.read_csv('/content/Titanic-Dataset.csv')
df.head(3)
````

````python
df=df.iloc[:,[2,5,6,7,9,1]]
df.head(3)
````

````python
df.dropna(inplace=True)
````

````python
df.isnull().sum()
````

#PCA manually

````python
X=df.iloc[:,:5]
y=df['Survived']
````

````python
#Setp-1:Standardization Scaling
from  sklearn.preprocessing import StandardScaler
ss=StandardScaler()
new_x=ss.fit_transform(X)
new_x
````

````python
#step2: Co-Varinace Matrix
cov=np.cov(new_x.T)
````

````python
#setp3: Eigen value and Eigen Vector
eig=np.linalg.eig(cov)
eig
````

````python
eig.eigenvalues
````

````python
#step4:Choose No. of Principle Component(PC)
W=eig.eigenvectors[:2,]
W
````

````python
#setp5: Projection of Scaled Data on PCs
pca_matrix=np.dot(new_x,np.transpose(W))
pca_matrix
````

````python
New_df=pd.DataFrame(pca_matrix,columns=['pc1','pc2'])
New_df['Survived']=y
New_df.head()


#5D to 2D te converted
````
